```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# This is a sample based on https://towardsdatascience.com/how-to-build-a-simple-recommender-system-in-python-375093c3fb7d
```

```python
movies = pd.read_csv("./movies.csv")
```

```python
movies.head(5)
```

```python
ratings = pd.read_csv("./ratings.csv")
```

```python
ratings.head(5)
```

```python
ratings = pd.merge(ratings, movies, on='movieId')[['userId', 'rating', 'title']]
```

```python
ratings.head(5)
```

```python
movie_ratings = pd.DataFrame(ratings.groupby('title')['rating'].mean())
```

```python
movie_ratings['number_of_ratings'] = ratings.groupby('title')['rating'].count()
```

```python
movie_ratings.head(5)
```

```python
movie_matrix = ratings.pivot_table(index='userId', columns='title', values='rating')
```

```python
movie_matrix.head(5)
```

```python
movie_ratings.sort_values('number_of_ratings', ascending=False)
```

```python
AFO_user_rating = movie_matrix['Air Force One (1997)']
contact_user_rating = movie_matrix['Contact (1997)']
```

```python
similar_to_air_force_one=movie_matrix.corrwith(AFO_user_rating)
```

```python
similar_to_air_force_one.head(20)
```

```python
similar_to_air_force_one.sort_values(ascending=False)
```

```python
similar_to_contact = movie_matrix.corrwith(contact_user_rating)
```

```python
corr_contact = pd.DataFrame(similar_to_contact, columns=['Correlation'])
corr_contact.dropna(inplace=True)
corr_contact.head()
```

```python
corr_AFO = pd.DataFrame(similar_to_air_force_one, columns=['correlation'])
corr_AFO.dropna(inplace=True)
corr_AFO.sort_values('correlation', ascending=False).head()
```

```python
corr_AFO = corr_AFO.join(movie_ratings['number_of_ratings'])
```

```python
corr_AFO.head(5)
```

```python
corr_contact = corr_contact.join(movie_ratings['number_of_ratings'])
```

```python
corr_contact.head(5)
```

```python
corr_AFO[corr_AFO['number_of_ratings'] > 100].sort_values(by='correlation', ascending=False).head(10)
```
